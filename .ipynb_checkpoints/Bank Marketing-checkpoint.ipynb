```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('whitegrid')

from scipy import stats

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

```

```python
data = pd.read_csv("../data/bank-additional-full.csv", sep=';')
```

```python
data.head()
```

```python
data.info()
```

```python
# Get a count of the distinct values for each column
data.nunique(axis=0)
```

```python
# Examine the label (Predicting whether a customer do a new deposit)
data['y'].value_counts()
```

```python
# Change Label to Numeric

def new_deposit(status):
    if status == 'yes':
        return 1
    elif status == 'no':
        return 0
    else:
        return 11111111
    
data['y'] = data['y'].apply(new_deposit)

data['y'].value_counts()   # Confirm there was no errors in the conversion
```

```python
# Check Distributions of numeric features

num_features = []
for column in data:
    if data[column].dtypes == 'int64':
        num_features.append(column)
    elif data[column].dtypes == 'float64':
        num_features.append(column)
        
num_features_df = data[num_features]

fig = plt.figure(figsize=(15, 20))
plt.suptitle('Numerical Features Distributions', fontsize=20)
for i in range(1, num_features_df.shape[1] + 1):
    plt.subplot(4, 3, i)
    f = plt.gca()
    f.axes.get_yaxis().set_visible(False)
    f.set_title(num_features_df.columns.values[i - 1])

    #vals = np.size(num_features_df.iloc[:, i - 1].unique())   # Use this for a higher level of granularity
    vals = 25
    
    plt.hist(num_features_df.iloc[:, i - 1], bins=vals, color='green', edgecolor='black')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
# Examine Correlations between numeric features

plt.figure(figsize=(15,10))
sns.heatmap(num_features_df.corr(), annot=True, linecolor='black', linewidths=1.5)
plt.title('Numerical Features Correlation')
plt.show()
```

```python
# Look at correlations in relation to whether or not a customer will deposit new money

corr_df = num_features_df.corr()['y'].sort_values(ascending=True)
plt.figure(figsize=(14,6))
corr_df[:-1].plot(kind='bar', color='grey', edgecolor='black', linewidth=2)
plt.title('Correlation with depositing new money', size=20)
plt.xlabel('Features', size = 15)
plt.ylabel('Correlation', size=15)
plt.xticks(rotation= 45, size=10)
plt.show()
```

```python
# Check distributions for categorical variables

cat_features = []
for column in data:
    if data[column].dtypes == object:
        cat_features.append(column)

cat_features_df = data[cat_features]     # Creates a dataframe of just categorical columns

fig = plt.figure(figsize=(15, 40))
plt.suptitle('Categorical Features Distributions', fontsize=30)
for i in range(1, cat_features_df.shape[1] + 1):
    plt.subplot(10, 2, i)
    f = plt.gca()
    f.axes.get_yaxis().set_visible(False)
    f.set_title(cat_features_df.columns.values[i - 1], fontsize=20)
   
    values = cat_features_df.iloc[:, i - 1].value_counts().values
    index = cat_features_df.iloc[:, i - 1].value_counts().index
    plt.pie(values, labels = index, autopct='%1.1f%%')
    plt.axis('equal')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
# Pdays is previous number of days since last contacted.  999 indicates they have never been contacted.
data[data['pdays'] != 999]['pdays'].value_counts().plot(kind='bar')
```

```python
# Convert the value of records where customer has never been contacted to prevent skewing the data
def been_contacted(days):
    if days == 999:
        return(-1)
    elif days != 999:
        return days

data['pdays'] = data['pdays'].apply(been_contacted)
```

```python
# Create a new feature of whether or not a customer has been contacted before or not form the 'pdays' feature
def contact_history(days):
    if days == -1:
        return(0)
    elif days >= 0:
        return(1)

data['prior_contact'] = data['pdays'].apply(contact_history)   # Create new feature

cat_features.append('prior_contact')             # Append new feature to list of categorical features
```

```python
# Remove Label from list of numerical features
num_features.remove('y')
```

```python
# Keep a list of all feature names
col_names = data.columns.values.tolist()

# Create label column
label = data['y']

# Drop the label column
data = data.drop('y', axis=1)
```

```python
# For now treat the unknown answers as if they could be an indicator of something.
transformation_pipeline = ColumnTransformer([
    ('num_scaling', StandardScaler(), num_features),
    ('cat_ohe', OneHotEncoder(), cat_features)
], remainder='passthrough')
```

```python
pipeline = Pipeline(steps=[
    ('preprocess', transformation_pipeline)
])
```

```python
pd.DataFrame(pipeline.fit_transform(data))
```
