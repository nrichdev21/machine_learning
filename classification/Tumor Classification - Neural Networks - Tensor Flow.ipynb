```python
X = df.drop('benign_0__mal_1', axis=1).values
```

```python
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train) # Keep in mind only want to fit to our training data
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model= Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

# Binary Classification - So use sigmoid for activation function
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test)) # High number of epochs to show overfitting example
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
# Tells us we are overfitting to our dataset, too many epochs.
```

```python
model= Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

# Binary Classification - So use sigmoid for activation function
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode = 'min', patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=1000, validation_data=(X_test, y_test),
         callbacks=[early_stop])   # Demonstratings early stop
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
# Can also add in drop out layers to prevent over fitting
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model= Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))       # Percentage of Neurons randomly turning off - between 0.2 and 0.5 is norm
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5)) 

# Binary Classification - So use sigmoid for activation function
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=1000, validation_data=(X_test, y_test),
         callbacks=[early_stop])   # Demonstratings early stop
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
